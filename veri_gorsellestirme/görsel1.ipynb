## Çaprazlamalar

```python
#çaprazlama veri seti içerisindeki yer alan değerlerin birlikte değerlendirilmesi demek.
import seaborn as sns
from pandas.api.types import CategoricalDtype
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
cut_kategoriler = ["Fair","Good","Very Good","Premium","Ideal"]
df.cut = df.cut.astype(CategoricalDtype(categories = cut_kategoriler, ordered = True))
df.head()
```

```python
sns.catplot(x = "cut", y = "price", data = df);
#cut ve price değişkenini birbirlerine göre çaprazlayarak görselleştirmiş olduk.
```

```python
#Burada kaliteler arttıkça her bir kategori sınıfında yoğunlaşmalar gözlenmektedir.
```

```python
sns.barplot(x = "cut", y = "price", hue = "color", data = df);
#boyut olarak color kategorik değişkenini ekledik.
```

#Daha fazla açıklanabilirlik elde ettik.Boyut arttıkmak bu işe yaradı.Buradaki price eksenindeki değerler iki gruptaki değişken bir arada bulunduğunda iki gruopby işlemi sonrasında price değişkeninin ortalamsı.
#Üzerine koyulan çubuklar ise standart sapma değerleri(Diğer bir açıdan standart hatalar olarak da değerlendirebiliriz).

```python
df.groupby(["cut","color"])["price"].mean()
#ilk önce cut'a göre kır sonra colora göre. Daha sonra price'a göre ortalama alma işlemi yap.
```
